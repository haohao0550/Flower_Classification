## Ki·ªÉm tra k√≠ch th∆∞·ªõc ·∫£nh sau khi crawl

```python
import random
import PIL.Image as Image
import os
folder = "Vietnam_flower"
image_paths = []

# L·∫•y t·∫•t c·∫£ ƒë∆∞·ªùng d·∫´n ·∫£nh t·ª´ c√°c folder con
for species_folder in os.listdir(folder):
    folder_path = os.path.join(folder, species_folder)
    if os.path.isdir(folder_path):
        img_files = [f for f in os.listdir(folder_path) if f.lower().endswith(('.jpg', '.jpeg', '.png', '.bmp', '.gif', '.webp'))]
        for f in img_files:
            image_paths.append(os.path.join(folder_path, f))

# Ch·ªçn 20 ·∫£nh ng·∫´u nhi√™n
sample_images = random.sample(image_paths, min(20, len(image_paths)))

# In k√≠ch th∆∞·ªõc c·ªßa 20 ·∫£nh random
for img_path in sample_images:
    try:
        with Image.open(img_path) as img:
            print(f"{img_path}: {img.width}x{img.height}")
    except Exception as e:
        print(f"L·ªói m·ªü {img_path}: {e}")
```

## B·ªè nh·ªØng ·∫£nh d∆∞·ªõi 224 x 224

```python
import os
from PIL import Image

root = "Vietnam_flower"

image_size = 224

count = 0
small_images = []

for species_folder in os.listdir(root):
    folder_path = os.path.join(root, species_folder)
    if os.path.isdir(folder_path):
        for file in os.listdir(folder_path):
            if file.lower().endswith(('.jpg', '.jpeg', '.png', '.webp')):
                img_path = os.path.join(folder_path, file)
                try:
                    with Image.open(img_path) as img:
                        if img.width < image_size and img.height < image_size:
                            count += 1
                            small_images.append((img_path, img.width, img.height))
                            print(f"üìå {img_path} ‚Äî {img.width}x{img.height}")
                except Exception as e:
                    print(f"‚ö†Ô∏è L·ªói m·ªü {img_path}: {e}")

print(f"\nüéâ T·ªïng s·ªë ·∫£nh nh·ªè h∆°n {image_size}x{image_size}px:", count)
```

## Resize image 224 x 224

```python
from PIL import Image
import os
from concurrent.futures import ThreadPoolExecutor

src_root = "Vietnam_flower"
dst_root = "../Dataset"
os.makedirs(dst_root, exist_ok=True)

def resize_image(src_img_path, dst_img_path):
    try:
        img = Image.open(src_img_path)
        img = img.resize((224, 224), Image.LANCZOS)
        img.save(dst_img_path)
    except Exception as e:
        print(f"L·ªói resize {src_img_path}: {e}")

for species_folder in os.listdir(src_root):
    src_folder_path = os.path.join(src_root, species_folder)
    dst_folder_path = os.path.join(dst_root, species_folder)
    if os.path.isdir(src_folder_path):
        os.makedirs(dst_folder_path, exist_ok=True)
        jobs = []
        with ThreadPoolExecutor(max_workers=16) as executor:  # TƒÉng s·ªë lu·ªìng n·∫øu m√°y m·∫°nh
            for file in os.listdir(src_folder_path):
                if file.lower().endswith((".jpg", ".png")):
                    src_img_path = os.path.join(src_folder_path, file)
                    dst_img_path = os.path.join(dst_folder_path, file)
                    jobs.append(executor.submit(resize_image, src_img_path, dst_img_path))
            for job in jobs:
                job.result()
```

## Split Folder

```python
import os
import shutil
import math

source_root = "../Dataset"   # th∆∞ m·ª•c dataset ban ƒë·∫ßu
output_roots = ["chunk/part1", "chunk/part2", "chunk/part3", "chunk/part4", "chunk/part5", "chunk/part6"]  # 6 th∆∞ m·ª•c ƒë√≠ch

# T·∫°o th∆∞ m·ª•c output n·∫øu ch∆∞a c√≥
for out in output_roots:
    os.makedirs(out, exist_ok=True)

# L·∫•y danh s√°ch c√°c th∆∞ m·ª•c con (v√≠ d·ª• 55 lo√†i hoa)
subfolders = sorted([
    f for f in os.listdir(source_root)
    if os.path.isdir(os.path.join(source_root, f))
])

total = len(subfolders)
print("T·ªïng s·ªë th∆∞ m·ª•c con:", total)

# S·ªë th∆∞ m·ª•c chia ra m·ªói ph·∫ßn
chunk_size = math.ceil(total / len(output_roots))

# Chia ƒë·ªÅu th√†nh 3 ph·∫ßn
parts = [
    subfolders[i:i + chunk_size]
    for i in range(0, total, chunk_size)
]

# Copy th∆∞ m·ª•c
for part_idx, part in enumerate(parts):
    target_root = output_roots[part_idx]
    print(f"\nüìÅ Copy v√†o {target_root}: {len(part)} th∆∞ m·ª•c")

    for folder in part:
        src = os.path.join(source_root, folder)
        dst = os.path.join(target_root, folder)
        shutil.copytree(src, dst)

print("\nüéâ Ho√†n t·∫•t chia dataset th√†nh 3 ph·∫ßn!")
```
